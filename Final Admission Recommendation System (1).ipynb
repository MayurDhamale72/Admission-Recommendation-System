```python
# Data preprocessing
le_category = LabelEncoder()
df['category_encoded'] = le_category.fit_transform(df['category'])
```

```python
#select features
X = df[['percentile', 'category_encoded']]
y_college = df['college_name']
y_branch = df['branch']
```

```python
# Train-test split
X_train, X_test, y_college_train, y_college_test, y_branch_train, y_branch_test = train_test_split(
    X, y_college, y_branch, test_size=0.3, random_state=42)
```

```python
# Train a Random Forest classifier for college names
college_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
college_classifier.fit(X_train, y_college_train)
```

```python
# Train a Random Forest classifier for branch names
branch_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
branch_classifier.fit(X_train, y_branch_train)
```

```python
# Predictions on the test set
college_predictions_test = college_classifier.predict(X_test)
branch_predictions_test = branch_classifier.predict(X_test)
```

```python
# User input
user_percentage = float(input("Enter your percentage: "))
user_category = input("Enter your category: ")
```

```python
# Encode user input category
user_category_encoded = le_category.transform([user_category])[0]
```

```python
# Make predictions with probabilities
college_probabilities = college_classifier.predict_proba([[user_percentage, user_category_encoded]])[0]
branch_probabilities = branch_classifier.predict_proba([[user_percentage, user_category_encoded]])[0]
```

```python
# Get top 5 indices with highest probabilities for college and branch
top_college_indices = college_probabilities.argsort()[-5:][::-1]
top_branch_indices = branch_probabilities.argsort()[-5:][::-1]
```

```python
# Get top 5 college and branch names
top_college_names = df['college_name'].unique()[top_college_indices]
top_branch_names = df['branch'].unique()[top_branch_indices]
```

```python
# Filter branches based on the suggested college
filtered_branches = df[df['college_name'].isin(top_college_names)]['branch'].unique()
```
